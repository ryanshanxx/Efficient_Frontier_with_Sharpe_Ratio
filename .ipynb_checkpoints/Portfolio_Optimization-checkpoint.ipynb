```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import quandl
%matplotlib inline
```

```python
# get adjusted closing prices of 27 selected companies with Quandl from 9 industries
quandl.ApiConfig.api_key = 'YxoLj8ezeNxXnnB4-oKX'
Communication_Services = ['T','VZ','CMCSA']
Consumer_Cyclical = ['AMZN','HD','DIS']
Consumer_Defensive = ['WMT','PG','KO']
Energy = ['XOM','BP','CVX']
Financial_Services = ['BAC','JPM','V']
Healthcare = ['JNJ','UNH','MRK']
Industrials = ['BA','HON','UNP']
Real_Estate = ['SPG','CCI','PLD']
Technology = ['MSFT','AAPL','GOOGL']

selected = Communication_Services + Consumer_Cyclical + Consumer_Defensive + Energy + Financial_Services + Healthcare + Industrials + Real_Estate + Technology

data = quandl.get_table('WIKI/PRICES', ticker = selected,
                        qopts = { 'columns': ['date', 'ticker', 'adj_close'] },
                        date = { 'gte': '2014-1-1', 'lte': '2017-1-1' }, 
                        paginate=True)
```

```python
clean = data.set_index('date')
table = clean.pivot(columns='ticker')
table.head()
```

```python
# read sp500 index ('Adj Close') from '2014-1-1' to '2017-1-1'
sp500 = pd.DataFrame(pd.read_csv('^GSPC_Project.csv', parse_dates=True, index_col='Date')['Adj Close'])
```

```python
sp500.head()
```

```python
stock_ret_daily = table.pct_change().dropna()
sp500_ret_daily = sp500.pct_change().dropna()
```

```python
slope = []
intercept = []
rvalue = []
pvalue = []
stderr = []
for i in selected:
    y = stock_ret_daily['adj_close'][i]-0.0017
    x = sp500_ret_daily['Adj Close']-0.0017
    lm = stats.linregress(x,y)
    slope.append(lm[0])
    intercept.append(lm[1])
    rvalue.append(lm[2])
    pvalue.append(lm[3])
    stderr.append(lm[4])
```

```python
stock_ret_annual = stock_ret_daily.mean() * 250
stock_vol_annual = np.sqrt(stock_ret_daily.var()) * np.sqrt(250)
sharpe_ratio = stock_ret_annual / stock_vol_annual
```

```python
d = {'slope':slope, 'intercept':intercept, 'rvalue':rvalue, 
     'pvalue':pvalue, 'stderr':stderr,'sharpe_ratio':sharpe_ratio}
stat = pd.DataFrame(d)
stat.index = selected
stat.head()
```

```python
new_selected = stat.sort_values(by='intercept', ascending=False)[:5].index
new_selected
```

```python
new_table_daily = pd.DataFrame()
for i in new_selected:
    new_table_daily[i] = stock_ret_daily['adj_close'][i]
new_table_daily.head()
```

```python
cov_annual = new_table_daily.cov() * 250
new_ret_annual = new_table_daily.mean() * 250
```

```python
# empty lists to store returns, volatility and weights of imiginary portfolios
port_returns = []
port_volatility = []
sharpe_ratio = []
stock_weights = []

# set the number of combinations for imaginary portfolios
num_assets = len(new_selected)
num_portfolios = 50000

#set random seed for reproduction's sake
np.random.seed(101)

# populate the empty lists with each portfolios returns,risk and weights
for single_portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    returns = np.dot(weights, new_ret_annual)
    volatility = np.sqrt(np.dot(weights.T, np.dot(cov_annual, weights)))
    sharpe = returns / volatility
    sharpe_ratio.append(sharpe)
    port_returns.append(returns)
    port_volatility.append(volatility)
    stock_weights.append(weights)
```

```python
# a dictionary for Returns and Risk values of each portfolio
portfolio = {'Returns': port_returns,
             'Volatility': port_volatility,
             'Sharpe Ratio': sharpe_ratio}
```

```python
# extend original dictionary to accomodate each ticker and weight in the portfolio
for counter,symbol in enumerate(new_selected):
    portfolio[symbol+' Weight'] = [Weight[counter] for Weight in stock_weights]
```

```python
# make a nice dataframe of the extended dictionary
df = pd.DataFrame(portfolio)
df.head()
```

```python
# get better labels for desired arrangement of columns
column_order = ['Returns', 'Volatility', 'Sharpe Ratio'] + [stock+' Weight' for stock in new_selected]

# reorder dataframe columns
df = df[column_order]
df.sort_values(by='Sharpe Ratio',ascending=False).head()
```

```python
# plot frontier, max sharpe & min Volatility values with a scatterplot
plt.style.use('seaborn-dark')
df.plot.scatter(x='Volatility', y='Returns', c='Sharpe Ratio',
                cmap='RdYlGn', edgecolors='black', figsize=(11, 8), grid=True)
plt.xlabel('Volatility (Std. Deviation)')
plt.ylabel('Expected Returns')
plt.title('Efficient Frontier')
plt.show()
```
